### 문제 1. 로그인 시스템 제어 (while & if-else)

사용자의 `id`와 `pw`를 입력받아 로그인을 처리하는 로직을 작성하세요.

- 시스템에 등록된 정보: `stored_id = "admin"`, `stored_pw = "pass123"`
- 사용자가 정확한 정보를 입력할 때까지 **무한히 반복**해서 입력을 받습니다.
- 로그인에 성공하면 `"환영합니다, 관리자님!"`을 출력하고 반복문을 탈출합니다.
- 로그인에 실패하면 `"아이디 또는 비밀번호가 틀렸습니다. 다시 시도하세요."`를 출력합니다.

```python
stored_id = "admin"
stored_pw = "pass123"
# 여기에 코드를 작성하세요.
```

````python
stored_id = "admin"
stored_pw = "pass123"

while True:
    user_id = input("아이디를 입력하세요: ")
    user_pw = input("비밀번호를 입력하세요: ")
    
    if user_id == stored_id and user_pw == stored_pw:
        print("환영합니다, 관리자님!")
        break
    else:
        print("아이디 또는 비밀번호가 틀렸습니다. 다시 시도하세요.")


 #✅ 코드 설명

# while True → 무한 반복
# 사용자가 아이디와 비밀번호 입력
# and 조건으로 아이디와 비밀번호 모두 일치하는지 확인
# 맞으면 break로 반복 종료 (아니면 무한루프)
# 틀리면 안내 메시지 출력 후 다시 입력
````

````python
stored_id = "admin"
stored_pw = "pass123"

while True:
    user_id = input("아이디: ")
    user_pw = input("비밀번호: ")

    if user_id == stored_id and user_pw == stored_pw:
        print("환영합니다, 관리자님!")
        break   # ← 이게 핵심!
    else:
        print("아이디 또는 비밀번호가 틀렸습니다. 다시 시도하세요.")

print("로그인 완료")
````

### 문제 2. 3-6-9 게임기 (for & range & if)

1부터 30까지의 숫자를 출력하되, 숫자에 **3, 6, 9**가 포함되어 있는 경우 숫자 대신 "짝!"을 출력하는 코드를 작성하세요.

- 팁: 숫자를 문자열로 변환하여 '3', '6', '9'가 들어있는지 확인(in 연산자)하면 쉽습니다.

```python
# 여기에 코드를 작성하세요.
```

````python
for i in range(1, 31):
    if '3' in str(i) or '6' in str(i) or '9' in str(i):
        print("짝!")
    else:
        print(i)


# 코드 설명
# range(1, 31) → 1부터 30까지 반복
# str(i) → 숫자를 문자열로 변환
# '3' in str(i) → 숫자 안에 3이 포함되어 있는지 확인
# 하나라도 포함되면 "짝!" 출력
````

### 문제 3. 장바구니 품목 필터링 및 할인 적용 (Comprehension)

다음은 장바구니에 담긴 상품 리스트입니다. 

가격이 **50,000원 이상인 상품만 골라 10% 할인**을 적용한 새로운 딕셔너리 `discounted_cart`를 만드세요.

- 결과 예시: `{'Laptop': 900000.0, 'Monitor': 180000.0}`
- 반드시 **딕셔너리 컴프리헨션**을 사용하여 한 줄로 작성하세요.

```python
cart = {"Laptop": 1000000, "Mouse": 15000, "Monitor": 200000, "Pad": 5000}
# 여기에 코드를 작성하세요.
```

````python
# 정답
cart = {"Laptop": 1000000, "Mouse": 15000, "Monitor": 200000, "Pad": 5000}

discounted_cart = {k: v * 0.9 for k, v in cart.items() if v >= 50000}

print(discounted_cart)


# 컴프리헨션 부분은 개인적인 공부가 더 필요하다.
# 필터링 조건은 if문으로 작성한다 if v >= 500000
````

### 문제 4. 구구단 가로 출력 (Nested Loop)

중첩 `for`문을 사용하여 구구단 2단부터 9단까지 출력하세요. 단, 각 단은 아래 예시처럼 **가로로 한 줄씩** 출력되어야 합니다.

- 출력 예시:
    
    `2*1=2 2*2=4 ... 2*9=18`
    
    `3*1=3 3*2=6 ... 3*9=27`
    

```python
# 여기에 코드를 작성하세요.
```

````python
# 정답 (chatGPT 도움)

for dan in range(2, 10):          # 2단 ~ 9단
    for num in range(1, 10):      # 1 ~ 9
        print(f"{dan}*{num}={dan*num}", end=" ")
    print()  # 한 단이 끝나면 줄바꿈
````

````python
for i in range(2, 10):
    for j in range(1, 10):
        print(f"{i}*{j}={i*j}", end="\t") # 탭 공백으로 가로 정렬
    print() # 한 단이 끝나면 줄바꿈
````

### 문제 5. 데이터 구조 재구성 (zip & Unpacking)

웹사이트 방문자 데이터가 `[(날짜, 방문자수), ...]` 형태의 튜플 리스트로 저장되어 있습니다. 이 데이터를 시각화하기 위해 날짜 리스트(`dates`)와 방문자수 리스트(`counts`)로 각각 분리(Unzipping)한 뒤, 방문자 수의 **평균**을 계산하세요.

```python
visitor_data = [("2026-03-01", 120), ("2026-03-02", 150), ("2026-03-03", 90), ("2026-03-04", 200)]
# 여기에 코드를 작성하세요.
```

````python
# 핵심은 zip + 언패킹(Unpacking) 입니다.
# chatGPT 도움

visitor_data = [("2026-03-01", 120), 
                ("2026-03-02", 150), 
                ("2026-03-03", 90), 
                ("2026-03-04", 200)]

# 1️⃣ 언지핑 (분리)
dates, counts = zip(*visitor_data)

# 2️⃣ 평균 계산
average = sum(counts) / len(counts)

print("날짜 리스트:", list(dates))
print("방문자 수 리스트:", list(counts))
print("평균 방문자 수:", average)
````
