```python
def safe_average(data):
    # 1️⃣ 입력이 리스트인지 확인
    if not isinstance(data, list):
        return "입력은 리스트여야 합니다."
    
    numbers = []
    
    # 2️⃣ 숫자만 추출 (bool 제외)
    for item in data:
        if isinstance(item, bool):
            continue
        if isinstance(item, (int, float)):
            numbers.append(item)
    
    # 3️⃣ 유효한 숫자가 없으면 0 반환
    if len(numbers) == 0:
        return 0
    
    # 4️⃣ 평균 계산
    average = sum(numbers) / len(numbers)
    
    # 5️⃣ 소수점 둘째 자리 반올림
    return round(average, 2)

print(safe_average([10, 20, 30, 40, 50]))
print(safe_average([85, 'N/A', 90, None, 78, 92]))
print(safe_average([10, 20.5, '30', 40, 'hello']))
print(safe_average(['a', 'b', 'c', None]))
print(safe_average("not a list"))
print(safe_average([]))
```

```python
students = {
	101 : {'name' : '김철수', 'score' : 85},
	102 : {'name' : '이영희', 'score' : 92},
	103 : {'name' : '박민수', 'score' : 67},
	104 : {'name' : '정수진', 'score' : 55},
	105 : {'name' : '최동현', 'score' : 78}
}

# 여기에 함수를 작성하세요
def get_student_info(students, student_id):
    try:
        # 1️⃣ 문자열 ID를 정수로 변환
        student_id = int(student_id)
        
        # 2️⃣ 해당 ID 학생 정보 조회
        student = students[student_id]
        
    except ValueError:
        return "유효하지 않은 ID 형식입니다."
    
    except KeyError:
        return "해당 ID의 학생이 존재하지 않습니다."
    
    # 3️⃣ 등급 계산
    score = student['score']
    
    if score >= 90:
        grade = 'A'
    elif score >= 80:
        grade = 'B'
    elif score >= 70:
        grade = 'C'
    elif score >= 60:
        grade = 'D'
    else:
        grade = 'F'
    
    # 4️⃣ 결과 반환
    return f"이름: {student['name']}, 점수: {score}, 등급: {grade}"

# 테스트 코드(수정하지 마세요)
print(get_student_info(students, '102')) # 정상 조회
print(get_student_info(students, 104)) # 정수로 직접 전달
print(get_student_info(students, '999')) # 존재하지 않는 ID
print(get_student_info(students, 'abc')) # 잘못된 형식
```
