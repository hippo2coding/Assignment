#### 문제 1. 사용자 입력 데이터 처리

#### 어떤 설문조사 시스템에서 나이를 입력받았습니다. 입력받은 `age_input = "25"`를 이용해 10년 뒤의 나이를 계산하여 "10년 뒤 당신은 35세가 됩니다."라고 출력하는 코드를 작성하세요.

```python
print("10년 뒤 당신은 35세가 됩니다")
# 'age_input ="25"'를 어떻게 이용할까?
# "25"는 따옴표로 감싸져 있어 숫자가 아닌 문자열이다. int 를 사용해야 할 거 같다.
# 문자열 상태에서는 + 10 을 사용 할 수 없다.

# int로 변환
age_input = 25
age_after_10_years = int(age_input) + 10

print(f"10년 뒤 당신은 {age_after_10_years}세가 됩니다.")

# 근데 왜 F-string을 사용해야 할까?
# 1. 가독성, 2. 문자열안에서 연산이 사용 가능 3. 속도가 빠르다(아직 와닿지 않는다.)
print("10년 뒤 당신은", age_after_10_years, "세가 됩니다.")
```

#### 문제 2. 리스트의 중복 제거와 정렬

#### 다음 리스트에서 중복된 과일 이름을 제거하고, 가나다 순(오름차순)으로 정렬된 리스트를 출력하세요.

#### fruits = ["사과", "바나나", "사과", "포도", "바나나", "체리"]
#### 여기에 코드를 작성하세요.

```python
# 1. set으로 변환하여 중복을 제거하고, 다시 list로 변환한다.
# why? set 에서는 오름차순으로 정렬이 불가능하니까

fruits = ["사과", "바나나", "사과", "포도", "바나나", "체리"]
unique_fruits = list(set(fruits))
print(unique_fruits)

# 2. 가나다 순(오름차순)으로 정렬합니다.
unique_fruits.sort()
print(unique_fruits)
```

#### 문제 3. 딕셔너리 안전하게 조회하기

#### 아래와 같은 성적표 딕셔너리가 있습니다. 사용자가 **"과학"** 과목의 점수를 조회하려고 할 때, 에러가 발생하지 않도록 하고 데이터가 없을 경우 "해당 과목의 점수가 없습니다."라고 출력하는 코드를 작성하세요.

#### scores = {"국어": 90, "수학": 85, "영어": 95}
# 여기에 코드를 작성하세요.

```python
scores = {"국어": 90, "수학": 85, "영어": 95}

#방법 1. 대괄호 사용 (직관적이지만, 위험할 수 있음)
# print(scores['과학'])
#key 값이면 에러 발생


# 방법2. get() 메소드 사용 (안전한 프로그래밍)
print(scores.get('과학', '해당 과목의 점수가 없습니다.'))
# 방법2도 근데 내 마음에 들지 않음.
```

```python
# 방법3. (chatGPT 도움, 내가 만들고 싶었던 것)


score_dict = {
    "국어": 90,
    "영어": 85,
    "수학": 100
}

while True:
    subject = input("조회할 과목명을 입력하세요 (종료하려면 'q'): ")
    
    if subject == 'q':
        print("조회 종료")
        break
    
    score = score_dict.get(subject)
    
    if score is None:
        print("해당 과목의 점수가 없습니다.")
    else:
        print(f"{subject} 점수는 {score}점입니다.")
```

#### 문제 5. 빈 데이터 판별기(심화)

사용자로부터 입력받은 검색어 `search_keyword`가 비어있는지 확인하는 코드를 작성하세요.

- 검색어가 있으면: **" '검색어'를 검색합니다."** 출력
- 검색어가 비어있거나 공백만 있다면(strip 활용): **"검색어를 입력해주세요."** 출력

search_keyword = "  "
#### 예시: 공백만 입력된 경우 여기에 코드를 작성하세요.

```python
# chatpgt 도움 
# 핵심은 strip()으로 공백 제거 후 비어있는지 확인하는 것입니다.

search_keyword = "  "  # 예시: 공백만 입력된 경우

if search_keyword.strip() == "":
    print("검색어를 입력해주세요.")
else:
    print(f"'{search_keyword.strip()}'를 검색합니다.")
```
